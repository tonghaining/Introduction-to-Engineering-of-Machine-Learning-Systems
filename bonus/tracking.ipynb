# Grafana

One of the remote servers has grafana installed (ask instructors if you get this far and have time to work on this!)

You can connect to it via an ssh tunnel

`ssh -N -L 3001:localhost:3001 remote-mlops`

and then go to `localhost:3001` in your browser. Ask instructors for the login password.

Take a look at dashboards like `Kubernetes / Compute Resources / Cluster`, and see if you can identify peaks where the server started getting used as the workshop hands-on started.

The cluster is running prometheus which polls all the pods in the Kubernetes cluster. Grafana simply visualizes things from Prometheus.

# Task
Your task is simply to try out designing a query, for example you could try to filter top 10% latency for an endpoint over a 5 minute moving window.

Take a look at the docs https://grafana.com/docs/
